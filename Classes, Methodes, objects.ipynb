```python
class Person:                          #class with capital letter
    department = 'school of information'
    def set_name(self, new_name):      #Methods like set name must include Self
        self.name = new_name
    def set_location(self, new_location):
        self.location = new_location
                                       #then we can call function using dot notation
```

```python
person = Person()
person.set_name('chris brok')
person.set_location('USA')
print('{} lives in {} and his department is {}'.format(person.name, person.location, person.department))
```

```python
# map function to find - map(function, iterable, iterable)
store1 = [1.1,2.2,3.3,4.4]
store2 = [1.2,2.3,3.4,4.5]
cheapest = map(min, store1, store2)
cheapest  # this gives reference value(map object) which is lasy evaluation mapping
# we can do a for loop for listing the requested

for item in cheapest:
    print(item)
```

```python
people = ['Dr. Christopher Brooks', 'Dr. Kevyn Collins-Thompson', 'Dr. VG Vinod Vydiswaran', 'Dr. Daniel Romero']

def split_title_and_name(person):
    title = person.split() [0]
    lastname = person.split() [-1]

    return '{} {}' .format(title , lastname)
print(map(split_title_and_name, people)) # gives object
list(map(split_title_and_name, people))  # gives list
```

```python
# lambda function to make simple statements or calculations used in clearing data - 
# lambda variables : function then pass values

My_function = lambda a, b, c : a+b
My_function(1, 2, 3)
```

```python
people = ['Dr. Christopher Brooks', 'Dr. Kevyn Collins-Thompson', 'Dr. VG Vinod Vydiswaran', 'Dr. Daniel Romero']
list(map(lambda person: person.split()[0] + ' ' + person.split()[-1], people))
```

```python
my_list = []      # pre example
for number in range(0, 10):
    if number % 2 == 0:
        my_list.append(number)
my_list
```

```python
# example on List Comprehensions
# its like lambda short and for simple operations

my_list = [number for number in range(0, 10) if number % 2 == 0]
my_list
```

```python
# write all possible 4 digits cases from 2 list 
numb= '123456'
lcase= 'abcdef'

lst = [ a+b+c+d for a in numb for b in numb for c in lcase for d in lcase]
lst
```
