```python
# Boolean masking
import pandas as pd

df = pd.read_csv('resources/Admission_Predict.csv', index_col=0)
df.columns = [x.lower().strip() for x in df.columns]
df.head()
```

```python
# we want the students of chance of admit higher than 0.7
admit_mask = df['chance of admit'] > 0.7 # this will return boolean values
admit_mask.head()
```

```python
# we can use this mask to hide the data doesnt meet our condidion with .where() function
df.where(admit_mask).head() # so we have only the data that meet the condition which we wanna use
```

```python
# The raws masked still in the data, if we want to get rid of them we use the function dropna()
df.where(admit_mask).dropna().head()
```

```python
#the panda devs created a shorthand syntax which combines where and dropna, doing both at once, 
#and in typical fashion, the just to overloaded the indexing operator to do this.
df[df['chance of admit'] > 0.7].head()
```

```python
df['gre score'].head()
```

```python
df[['gre score','toefl score']].head()  # [[to get mor than one column]]
```

```python
df[df['gre score']>320].head() # each of these is mimicing functionality from either .loc() or .where() or .dropna()
```

```python
#let's talk about combining multiple Boolean masks, such as multiple criteria for including. 
#In bit masking and other places in computer science, this is done with "and". If both masks must be True, 
#for a True value, and to be in the final mask, or "or" if only one needs to be True. Unfortunately, 
#it doesn't feel quite as natural in pandas. For instance, if you want to take two Boolean series and add them together. 
(['chance of admit'] > 0.7) and (df['chance of admit'] < 0.9)
```

```python
#instead we must use 
(df['chance of admit'] > 0.7) & (df['chance of admit'] < 0.9)
```

```python
# A common error for new pandas users is to try and do Boolean comparisons using the ampersand operator, 
#but not putting parentheses around the individual terms you are interested in.
# in another way we can use functions
df['chance of admit'].gt(0.7) & df['chance of admit'].lt(0.9)
```

```python
#also we can short cut 
df['chance of admit'].gt(0.7).lt(0.9)
```

# Indexing DataFrames

```python
#Another option for setting an index is to use the set_index function. 
#This function takes a list of columns and promotes those columns to an index.
import pandas as pd

df = pd.read_csv('resources/Admission_Predict.csv', index_col=0)
df.head()
```

```python
#We can do this using the indexing operator on the string that has the column label, 
#then we can use the set index to set the index of the column to the chance of admit.
df['Serial Number'] = df.index # make a new column named serial number and put index values
#then we set the index to another column
df = df.set_index('Chance of Admit ')
df.head()
```

```python
#We can get rid of the index completely by calling the function reset_index. This promotes the index into a column, 
#and creates a default numbered index.
df = df.reset_index()
df.head()
```

```python
# One nice feature of Pandas is multi-level indexing. This is similar to composite keys in relational database systems. 
#To create a multi-level index, we simply call set_index and give it a list of columns that we're interested in promoting 
#to an index. 
df = pd.read_csv('resources/census.csv', index_col=0)

df.head()
```

```python
df = df.reset_index()

df.head()
```

```python
#there are two summarized levels; one that contains summary data for the whole country, and one that contains summary data 
#for each state. I want to see a list of all the unique values in a given column.
df['SUMLEV'].unique()
# to get the different values only
```

```python
#we want only 50 so
df = df[df['SUMLEV'] == 50]
df.head()
```

```python
#let's reduce the data that we're going to look at to just the total population estimates, and the total number of births. 
#We can do this by creating a list of column names that we want to keep then project those and assign the resulting DataFrame 
#to our df variable. 
keepcolumns = ['STNAME','CTYNAME','POPESTIMATE2010','POPESTIMATE2011','POPESTIMATE2012']
df = df[keepcolumns]
df.head()
```

```python
# to set multiple indexes we can call 2 columns in the set_index function
df = df.set_index(['STNAME','CTYNAME'])
df.head()
```

```python
#we wand data for michigan - washtenaw county
df.loc['Michigan','Washtenaw County']
```

```python
# If you are interested in comparing two counties, for example, Washtenaw and Wayne County, 
#we can pass a list of tuples describing the indices that we wish to query into the loc attribute
df.loc[ [('Michigan','Washtenaw County'),('Michigan','Wayne County'),('Alabama','Autauga County')] ]
```
